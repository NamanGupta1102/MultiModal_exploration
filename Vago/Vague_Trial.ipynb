```python
# Use a pipeline as a high-level helper
# from transformers import pipeline


from transformers import pipeline

classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
```

```python
# labels = ["Not vague","Approximation Vagueness", "Generality Vagueness", "Degree Vagueness", "Combinatorial Vagueness"]
labels = ["Vague","Not Vague"]
```

```python
word = "joke"
prompt = f"Determine the type of vagueness for the word '{word}'."

result = classifier(word, labels)
print(result)
```

```python
# labels = ["Not vague","Approximation Vagueness", "Generality Vagueness", "Degree Vagueness", "Combinatorial Vagueness"]
labels = ["Not Vague","Vague"]
```

```python
def vago(sen, classifier = classifier):
  sen = sen.lower()
  sen = sen.split()
  # print(sen)
  vag = 0
  total = len(sen)
  for i in sen:
    ans = classifier(i, labels)['scores']
    # print(ans)
    if ans[0]<ans[1]:
      vag+=1
  return (vag/total)


```

```python
sentence = "Take a left, walk 100m straight, climb up yellow stair case, take right from red sofa."
vago(sentence)
```

```python
## Imports
import h5py
import numpy as np
import os
```

```python
def decode_string(last_action_str):
    # Decode the bytes to string
    decoded_actions = []
    for action in last_action_str:
        # Remove null bytes and decode
        decoded = np.trim_zeros(action).tobytes().decode('utf-8')
        decoded_actions.append(decoded)
    print("Last action strings:")
    for i, action in enumerate(decoded_actions):  # Print first 5 actions
        print(f"{i}: {action}")
```
